```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [5,6,2]
x1 = [1,2,3]
y1 = [5,6,8]
plt.plot(x,y,label='first')
plt.plot(x1,y1,label='second')
plt.xlabel('its x axis')
plt.ylabel('its y axis')
plt.title('ITS TITLE \n 2nd line')
plt.legend()
plt.show()
```

```python
y = [20,10,15,10,11,17,20,21,22,24,28,30,20,10,11,17,12,14,18,50,40,41,47,34]
bins = [0,10,20,30,40,50,60]
plt.hist(y,bins,histtype='bar',rwidth=0.8)
plt.xlabel('age bins')
plt.ylabel('frequency')
plt.title('ITS TITLE \n 2nd line')
plt.legend()
plt.show()
```
