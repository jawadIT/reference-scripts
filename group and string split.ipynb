```python
shopstr = "apples,4,bananas,3,corn,1,dish soap,1"
prices = "apples,0.5,bananas,0.2,corn,0.25,dish soap,3"
coupons = "bananas,0.5,corn,0.1,dish soap,0.2"

def split_items(item):
    try:
        val = int(item)
        return val,'quantity'
    except ValueError:
        try:
            val = float(item)
            return val,'quantity'
        except ValueError:
            return int(0),'product'

def best_coupons(shopstr, prices, coupons):
    couponslst = coupons.split(",")
    priceslst = prices.split(",")
    shoplst = shopstr.split(",")

    prev = ''
    typ = ''
    dfp = pd.DataFrame()
    prlst = []
    qlst = []
    for i in range(0,len(shoplst)):
        qunt , typ = split_items(shoplst[i])
        if (typ == "quantity" and prev == "product"):
            prev = typ
            qlst.append(qunt)
        else:
            prev = typ
            prlst.append(typ)
    dfp['product'] = prlst
    dfp['quantity'] = qlst

    prev = ''
    typ = ''
    dfp = pd.DataFrame()
    prlst = []
    qlst = []
    for i in range(0,len(shoplst)):
        qunt , typ = split_items(shoplst[i])
        if (typ == "quantity" and prev == "product"):
            prev = typ
            qlst.append(qunt)
        else:
            prev = typ
            prlst.append(typ)
    dfp['product'] = prlst
    dfp['quantity'] = qlst


    prev = ''
    typ = ''
    dfpr = pd.DataFrame()
    prlst = []
    prilst = []
    for i in range(0,len(priceslst)):
        qunt , typ = split_items(priceslst[i])
        if (typ == "quantity" and prev == "product"):
            prev = typ
            prilst.append(qunt)
        else:
            prev = typ
            prlst.append(typ)

    dfpr['product'] = prlst
    dfpr['price'] = prilst


    prev = ''
    typ = ''
    dfc = pd.DataFrame()
    prlst = []
    clst = []
    for i in range(0,len(couponslst)):
        qunt , typ = split_items(couponslst[i])
        if (typ == "quantity" and prev == "product"):
            prev = typ
            clst.append(qunt)
        else:
            prev = typ
            prlst.append(typ)
    dfc['product'] = prlst
    dfc['coupon'] = clst

    dfp.set_index('product')
    dfpr.set_index('product')
    dfc.set_index('product')
    
    df = pd.concat([dfp,dfpr,dfc],axis=1)
    
    df.set_index('coupon')
    
    df.sort_index(inplace=True)
    
    strng = str(df['product'][1]+ ',' +  str(df['product'][2])
    
    return(strng)


print(best_coupons(shopstr, prices, coupons))
```
