```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_raw = pd.read_csv(r'C:\Users\jawad_zf1uaw5\Desktop\pers\data\AAPL.csv')
```

```python
df_raw
```

```python
df_filt = df_raw.iloc[:,1:2].values
```

```python
df_filt
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
```

```python
df_scaled = scaler.fit_transform(df_filt)
```

```python
df_scaled
```

```python
features = []
labels = []
for i in range(60,1259):
    features.append(df_scaled[i-60:i,0])
    labels.append(df_scaled[i,0])
```

```python
features , labels = np.array(features), np.array(labels)
```

```python
features.shape
```

```python
features = np.reshape(features, (features.shape[0],features.shape[1],1))
```

```python
features.shape
```

```python
from keras.models import Sequential 
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
model = Sequential()
model.add(LSTM(units=50,return_sequences=True,input_shape=(features.shape[1],1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.fit(features,labels,epochs=100,batch_size=32)
```

```python
df_test_raw = pd.read_csv(r'C:\Users\jawad_zf1uaw5\Desktop\pers\data\AAPLT.csv')
df_test_filter = df_test_raw.iloc[:, 1:2].values
df_merge = pd.concat((df_raw['Open'], df_test_raw['Open']), axis=0)
```

```python
test_inputs = df_merge[len(df_merge) - len(df_test_raw) - 60:].values
test_inputs = test_inputs.reshape(-1,1)
test_inputs = scaler.transform(test_inputs)
```

```python
test_features = []
for i in range(60, 80):
    test_features.append(test_inputs[i-60:i, 0])
```

```python
test_features = np.array(test_features)
test_features = np.reshape(test_features, (test_features.shape[0], test_features.shape[1], 1))
```

```python
predictions = model.predict(test_features)
```

```python
predictions = scaler.inverse_transform(predictions)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df_test_filter, color='blue', label='Actual Apple Stock Price')
plt.plot(predictions , color='red', label='Predicted Apple Stock Price')
plt.title('Apple Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Apple Stock Price')
plt.legend()
plt.show()
```
